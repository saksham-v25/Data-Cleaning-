```python
import pandas as pd
import numpy as np
```

```python
teams=pd.read_csv("/teams.csv")
```

```python
teams
```

```python
X = teams[["athletes", "prev_medals"]].copy()
y = teams[["medals"]].copy()
```

```python
X
```

```python
y
```

```python
X["intercept"] = 1.0
```

```python
X=X[["intercept","athletes","prev_medals"]]
```

```python
X
```

```python
X_T=X.T
```

```python
X_T
```

```python
B = np.linalg.inv(X_T @ X) @ X_T @ y
```

```python
B
```

```python
B.index = X.columns
```

```python
B
```

```python
predictions = X @ B
```

```python

predictions
```

```python

SSR = ((y - predictions) ** 2).sum()
```

```python

SSR
```

```python

TSS = ((y - y.mean()) ** 2).sum()
```

```python

TSS
```

```python

R2 = 1 - (SSR / TSS)
```

```python

R2
```

```python

from sklearn.linear_model import LinearRegression
```

```python

lr = LinearRegression()
```

```python

lr.fit(teams[["athletes", "prev_medals"]], teams[["medals"]])
```

```python

lr.intercept_
```

```python

lr.coef_
```
